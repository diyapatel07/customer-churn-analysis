```python
import pandas as pd
# Load dataset
df = pd.read_csv("C:/Users/DIYA/Downloads/customer_churn_dataset.csv")
print(df.shape)
df.head()
```

```python
# Check data types and missing
print(df.dtypes)
print(df.isnull().sum())

# Convert TotalCharges to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Quick EDA for Customer Churn

print("ðŸ”¹ Churn Distribution (%):")
print(df['Churn'].value_counts(normalize=True) * 100)
print("\n")

print("ðŸ”¹ Churn by Contract Type (%):")
print(pd.crosstab(df['Contract'], df['Churn'], normalize='index') * 100)
print("\n")

print("ðŸ”¹ Average Monthly Charges by Churn:")
print(df.groupby('Churn')['MonthlyCharges'].mean())
print("\n")

print("ðŸ”¹ Average Tenure by Churn:")
print(df.groupby('Churn')['tenure'].mean())
print("\n")

print("ðŸ”¹ Churn by Internet Service Type (%):")
print(pd.crosstab(df['InternetService'], df['Churn'], normalize='index') * 100)
print("\n")
```
